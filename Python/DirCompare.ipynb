```python
import os
import filecmp

def files_in_tree(root):
    """
    Return a dict mapping relative paths -> absolute paths
    for all files under root.
    """
    out = {}
    for dirpath, _, filenames in os.walk(root):
        for f in filenames:
            apath = os.path.join(dirpath, f)
            rpath = os.path.relpath(apath, root)
            out[rpath] = apath
    return out

def compare_trees(dirA, dirB):
    filesA = files_in_tree(dirA)
    filesB = files_in_tree(dirB)

    # Only files present in both directory trees
    common = set(filesA.keys()) & set(filesB.keys())

    different = []

    for rpath in sorted(common):
        if not filecmp.cmp(filesA[rpath], filesB[rpath], shallow=False):
            different.append(rpath)

    return different


# Example usage:
dirA = "/glade/work/juliob/GW_UnitTest/cases/xy-Baseline-v3.0"
dirB = "/glade/work/juliob/GW_UnitTest/cases/xy-vortsrc-01"


differences = compare_trees(dirA, dirB)

for f in differences:
    print("Different:", f)
```
